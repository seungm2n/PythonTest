```python
import pandas as pd

agg_prd_df = pd.DataFrame({
    'L' : [5141, 5365, 5084],
    'M' : [4968, 4977, 4915],
    'S' : [5012, 4991, 5029],
    'XL' : [5071, 4976, 5023],
    'XS' : [5195, 4954, 5088]
}, index= ['Jacket','Shirt','Trousers'])

agg_prd_df
```

```python
agg_prd_df.iloc[0, 1]
```

```python
agg_prd_df.iloc[0, 1] = 5000
agg_prd_df
```

```python
agg_prd_df.iloc[:, 1:3]
```

```python
agg_prd_df.loc['Jacket', 'M']
```

```python
agg_prd_df.loc['Shirt':'Trousers', 'L':'S']
```

```python
agg_prd_df.loc['Shirt':'Trousers', ['XL','S']]
```
