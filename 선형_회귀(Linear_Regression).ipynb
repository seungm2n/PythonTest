# 리니어 리그레션 모델 만들기

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y_train, X_train)
```

```python
model = model.fit()
```

```python
model.summary()
```

# 예측 및 평가

```python
model.predict(X_test)
```

```python
y_test
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```
